Question 4

```python
random_array=np.random.randint(low=35 ,high=50 ,size=(6,3))

print(random_array)

#using .T
transpose_random_array = random_array.T
print(transpose_random_array)

#using np.transpose()
transpose_random_array = np.transpose(random_array)
print(transpose_random_array)

#using .transpose()
transpose_random_array = random_array.transpose()
print(transpose_random_array)
```

```python
random_array1=np.random.randint(low=5 ,high=20 ,size=(2,2))
print(random_array1)

random_array2=np.random.randint(low=10 ,high=16 ,size=(2,2))
print(random_array2)

#dot product of two array
dot_product=np.dot(random_array1,random_array2)

print("The dot product of the two array is :")
print(dot_product)
```
